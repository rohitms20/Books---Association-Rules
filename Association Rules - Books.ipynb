# Association Rules:
- Association Rules refer to a data mining technique used to discover interesting relationships or associations between items in large datasets. The technique is based on the concept of frequent itemsets, which are sets of items that appear together in transactions above a certain threshold (minimum support).<br>
<br>
- The most common measure used to evaluate association rules is **support** and **confidence** - Support measures the frequency of occurrence of an itemset in the dataset, while confidence measures the strength of the association between two items in the itemset.<br>
<br>
- **The Apriori Algorithm** is a widely used algorithm for mining frequent itemsets and association rules. It generates a candidate itemset by joining pairs of frequent itemsets and prunes the itemset that does not meet the minimum support threshold. The process is repeated until no more frequent itemsets can be generated.

- Association rules have many practical applications, such as **market basket analysis**, where they are used to identify which products are frequently purchased together in a store. This information can then be used to make strategic decisions, such as product placement and promotion, to increase sales and customer satisfaction.

# Association Rules
Association rules refer to a data mining technique used to discover interesting relationships or associations between items in large datasets. The technique is based on the concept of frequent itemsets, which are sets of items that appear together in transactions above a certain threshold (minimum support).

The most common measure used to evaluate association rules is support and confidence. 
- **Support** measures the frequency of occurrence of an itemset in the dataset.
- **Confidence** measures the strength of the association between two items in the itemset.

**The Apriori Algorithm** is a widely used algorithm for mining frequent itemsets and association rules. It generates a candidate itemset by joining pairs of frequent itemsets and prunes the itemset that does not meet the minimum support threshold. The process is repeated until no more frequent itemsets can be generated.

Association rules have many practical applications, such as market basket analysis, where they are used to identify which products are frequently purchased together in a store. This information can then be used to make strategic decisions, such as product placement and promotion, to increase sales and customer satisfaction.

- Confidence(I1->I2)=Number of Transactions containing I1 and I2/(Total Number of Transactions containing I1)

- Lift: Lift is the ratio between the confidence and support expressed as :

- **Lift(I1->I2)=Confidence(I1->I2)/Support(I2)**

## Association rules with 10% Support and 30% confidence

```python
#With 10% Support
frequent_itemsets=apriori(data,min_support=0.1,use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

### Observations from Support value:

**Head 5**

- 42 percent of 100 purchases are ChildBks
- 24 percent of 100 purchases are YouthBks
- 43 percent of 100 purchases are CookBks
- 28 percent of 100 purchases are DoltYBks
- 21 percent of 100 purchases are RefBks

**Tail 5**

- 10 percent of 100 purchases are ChildBks, GeogBks, DoItYBks
- 10 percent of 100 purchases are ChildBks, GeogBks, ArtBks
- 10 percent of 100 purchases are ArtBks, CookBks, DoItYBks
- 10 percent of 100 purchases are GeogBks, CookBks, DoItYBks
- 10 percent of 100 purchases are GeogBks, ArtBks, CookBks

```python
#With 30% Confidence
rules = association_rules(frequent_itemsets,metric='confidence', min_threshold=0.3)
rules
```

### Observations
- Antecedent support variable tells us probability of antecedent products alone
- Consequents support variable tells us probability of consequents products alone
- The support value is the value of the two products (Antecedents and Consequents)
- Confidence is an indication of how often the rule has been found to be true.
- The ratio of the observed support to that expected if X and Y were independent.
- A leverage value of 0 indicates independence. Range will be [-1 1]
- A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]

```python
rules.sort_values('lift',ascending=False)
```

```python
#Lift Ratio > 1 is a good influential rule in selecting the associated transactions
lift=rules[rules.lift>1]
lift
```

## Visualization of obtained rule
### Scatterplot

```python
plt.figure(figsize=(16,9))
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
matrix = lift.pivot('antecedents','consequents','lift')
matrix
```

### Heatmap

```python
plt.figure(figsize=(20,6),dpi=250)
sns.heatmap(matrix,annot=True)
plt.title('HeatMap - ForLiftMatrix')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

### Barplot

```python
plt.figure(figsize=(20,6),dpi=250)
sns.barplot("support","confidence",data=lift)
plt.title("support vs confidence")
plt.show()
```

```python
plt.figure(figsize=(16,9))
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'], 
fit_fn(rules['lift']))
plt.xlabel('lift')
plt.ylabel('Confidence')
plt.title('lift vs Confidence')
```
